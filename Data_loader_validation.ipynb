```python
import data_loader as dl
import pandas as pd

def test_print_sample_text_and_text_corrected(parquet_text_file, parquet_text_corrected_file):
    # Load the parquet files
    text_df = pd.read_parquet(parquet_text_file)
    text_corrected_df = pd.read_parquet(parquet_text_corrected_file)

    # Merge the dataframes on 'l1' (and other columns if necessary) to ensure matching
    merged_df = pd.concat([text_df, text_corrected_df], axis=1)

    # Print 10 matched samples
    print("Matched samples of text and text_corrected:\n")
    for i in range(min(10, len(merged_df))):  # Print the first 10 rows
        print(f"Row {i + 1}:")
        print(f"Text          : {merged_df['cleaned_text'].iloc[i]}")
        print(f"Text Corrected: {merged_df['cleaned_text_corrected'].iloc[i]}")
        print(f"l1            : {merged_df['l1'].iloc[i]}")
        print("-" * 80)

def main():
    # File paths for the Excel files
    file_paths = ["Final database (main prompts).xlsx", "Final database (alternative prompts).xlsx"]

    # Generate the text and text_corrected parquet files using the new functions
    dl.get_clean_sentences_and_labels_text(*file_paths)
    dl.get_clean_sentences_and_labels_wo_split_new(*file_paths)

    # Test and print the samples
    test_print_sample_text_and_text_corrected('clean_text_nosplit.parquet', 'clean_corrected_nosplit.parquet')

if __name__ == "__main__":
    main()
```
